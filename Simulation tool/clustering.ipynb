# Clustering

# Affinity Propagation

```python
from sklearn.datasets import make_classification
from sklearn.cluster import AffinityPropagation
from matplotlib import pyplot
from matplotlib import pyplot
from numpy import where
from numpy import unique
import pandas as pd
tp_df = pd.read_excel("SimData_Transplant_Ver1.0.xls")
```

```python
model = AffinityPropagation(damping=0.8)
X = tp_df.drop(["Cens","Time"], axis=1)
Y = tp_df["Time"]
X = pd.get_dummies(X)
X = X.to_numpy()
model.fit(X)
yhat = model.predict(X)
# retrieve unique clusters
clusters = unique(yhat)
# create scatter plot for samples from each cluster
count = 0
for cluster in clusters:
    # get row indexes for samples with this cluster
    count+=1
    row_ix = where(yhat == cluster)
    print("Cluster: ",count, "Indexes", row_ix[0])
    # create scatter of these samples
    pyplot.scatter(X[row_ix, 0], X[row_ix, 1])
# show the plot
#print("Total Cluster")
pyplot.show()
```

```python
model = AffinityPropagation(damping=0.9) # use other hyperparameter, the cluster does not merge
X = tp_df.drop(["Cens","Time"], axis=1)
Y = tp_df["Time"]
X = pd.get_dummies(X)
X = X.to_numpy()
model.fit(X)
yhat = model.predict(X)
# retrieve unique clusters
clusters = unique(yhat)
# create scatter plot for samples from each cluster
count = 0
for cluster in clusters:
    # get row indexes for samples with this cluster
    count+=1
    row_ix = where(yhat == cluster)
    print("Cluster: ",count, "Indexes", row_ix[0])
    # create scatter of these samples
    pyplot.scatter(X[row_ix, 0], X[row_ix, 1])
# show the plot
#print("Total Cluster")
pyplot.show()
```

# Agglomerative Clustering

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hierarchical_model = AgglomerativeClustering(n_clusters=52)
yhat = hierarchical_model.fit_predict(X)
clusters = unique(yhat)
count = 0
# create scatter plot for samples from each cluster
for cluster in clusters:
    count+=1
    # get row indexes for samples with this cluster
    row_ix = where(yhat == cluster)
    print("Cluster: ",count, "Indexes", row_ix[0])
    # create scatter of these samples
    pyplot.scatter(X[row_ix, 0], X[row_ix, 1])
# show the plot
pyplot.show()
```

# Kmeans

```python
from sklearn.cluster import KMeans
km_model = KMeans(n_clusters=52, random_state = 42)
yhat = km_model.fit_predict(X)
clusters = unique(yhat)
count = 0
# create scatter plot for samples from each cluster
for cluster in clusters:
    count+=1
    # get row indexes for samples with this cluster
    row_ix = where(yhat == cluster)
    print("Cluster: ",count,"Indexes", row_ix[0])
    # create scatter of these samples
    pyplot.scatter(X[row_ix, 0], X[row_ix, 1])
# show the plot
pyplot.show()
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
db_model = DBSCAN(eps=20, min_samples=10)
yhat = db_model.fit_predict(X)
clusters = unique(yhat)
count = 0
# create scatter plot for samples from each cluster
for cluster in clusters:
    count+=1
    # get row indexes for samples with this cluster
    row_ix = where(yhat == cluster)
    print("Cluster: ",count)
    # create scatter of these samples
    pyplot.scatter(X[row_ix, 0], X[row_ix, 1])
# show the plot
pyplot.show()
```

# Gaussian Mixture Model

```python
from sklearn.mixture import GaussianMixture
GM_model = GaussianMixture(n_components=52,random_state = 42)
yhat = GM_model.fit_predict(X)
clusters = unique(yhat)
count = 0
# create scatter plot for samples from each cluster
for cluster in clusters:
    count+=1
    # get row indexes for samples with this cluster
    row_ix = where(yhat == cluster)
    print("Cluster: ",count)
    # create scatter of these samples
    pyplot.scatter(X[row_ix, 0], X[row_ix, 1])
# show the plot
pyplot.show()
```
